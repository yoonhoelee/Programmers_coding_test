# 문제 설명
네트워크란 컴퓨터 상호 간에 정보를 교환할 수 있도록 연결된 형태를 의미합니다. 예를 들어, 컴퓨터 A와 컴퓨터 B가 직접적으로 연결되어있고, 컴퓨터 B와 컴퓨터 C가 직접적으로 연결되어 있을 때 컴퓨터 A와 컴퓨터 C도 간접적으로 연결되어 정보를 교환할 수 있습니다. 따라서 컴퓨터 A, B, C는 모두 같은 네트워크 상에 있다고 할 수 있습니다.

컴퓨터의 개수 n, 연결에 대한 정보가 담긴 2차원 배열 computers가 매개변수로 주어질 때, 네트워크의 개수를 return 하도록 solution 함수를 작성하시오.

# 제한사항
컴퓨터의 개수 n은 1 이상 200 이하인 자연수입니다.
각 컴퓨터는 0부터 n-1인 정수로 표현합니다.
i번 컴퓨터와 j번 컴퓨터가 연결되어 있으면 computers[i][j]를 1로 표현합니다.
computer[i][i]는 항상 1입니다.

# 입출력 예
n = 3
computers = [[1, 1, 0], [1, 1, 0], [0, 0, 1]]
return = 2

n = 3
computers = [[1, 1, 0], [1, 1, 1], [0, 1, 1]]
return = 1

# 입출력 예 설명
예제 #1
아래와 같이 2개의 네트워크가 있습니다.


예제 #2
아래와 같이 1개의 네트워크가 있습니다.


# 출처
https://programmers.co.kr/learn/courses/30/lessons/43162?language=python3

```python
def solution(n, computers):
    answer = 0
    bfs = []
    visited = [0]*n
    while 0 in visited:
        bfs.append(visited.index(0))
        while bfs:
            node = bfs.pop(0)
            visited[node] = 1
            for i in range(n):
                if visited[i] == 0 and computers[node][i] == 1:
                    bfs.append(i)
        answer += 1
    return answer

solution(3, [[1, 1, 0], [1, 1, 0], [0, 0, 1]])
```

# 풀이
(1) 먼저, 탐색을 위한 큐, 방문한 노드를 체크해 둘 리스트, 네트워크의 개수를 저장할 변수를 생성합니다.

(2) 탐색을 시작할 첫 노드를 큐에 넣습니다. 큐에 들어가는 값은 인덱스로, 이 문제에서는 computers 리스트의 첫번째 인덱스를 큐에 넣는 것으로 시작합니다. 그리고 첫 노드를 방문했음을 표시합니다.

(3) 큐가 빌 때까지 반복문을 아래 내용을 반복으로 수행합니다.

(3-1) 큐의 앞에서부터 노드를 하나씩 꺼냅니다.

(3-2) 꺼낸 노드의 인접 노드들을 방문하는 반복문을 수행합니다.

(3-2-1) 인접 노드가 방문된 적이 없다면 큐에 넣고, 방문했음을 표시합니다.

# 느낀점
역시 레벨 3의 문제여서 그런지 문제가 매우 어려웠다. DFS 보다는 BFS의 개념을 사용하는 것이 맞다고 판단을 하였으나, 코드로 어떻게 구현을 할지에 대해서는 생각이 나지 않았다. BFS와 DFS에 대해서도 보다 더 깊게 공부를 해봐야할 것 같다.
